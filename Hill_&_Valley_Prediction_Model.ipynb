# **Import Library**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Lorddhaval/Dataset/main/Hill%20Valley%20Dataset.csv')
```

```python
y = df['Class']
```

```python
y.shape
```

```python
X = df.drop('Class', axis=1)
```

```python
X.shape
```

```python
X
```

# **Get Plot of First Two Rows**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X.iloc[0,:])
plt.title('Valley');
```

```python
plt.plot(X.iloc[1,:])
plt.title('Hill');
```

# **Get X Variable Standardized**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
X.shape
```

# **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify = y, random_state=222529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# *Get Model Train*

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

# **Get Model Prediction**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
lr.predict_proba(X_test)
```

# **Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
